```python
# specify input and output files here
file_input = '<input file>'
file_output = '<output file>'
```

```python
import pandas as pd

# parse and rename columns
# given format works for day-ahead prices ("Großhandelspreise") downloaded from https://www.smard.de/home/downloadcenter/download-marktdaten
df = pd.read_csv(file_input, sep=';', header=0, names=['time', 'price'], usecols=[0, 2], decimal=',')

# convert time column to proper format
df['time'] = pd.to_datetime(df['time'], format='%d.%m.%Y %H:%M')
```

```python
# correct for daylight saving time

# missing data point when switching to daylight saving time:
# 2114 2025-03-30 03:00:00
# debug with:
#   df.loc[2112:2115, 'time']

# duplicate data point when switching back to standard time
# 7153 2025-10-26 02:00:00
# debug with:
#   df['time'][df['time'].duplicated()]
#   df.loc[7152:7155, 'time']

df.loc[2114:7153, 'time'] = df.loc[2114:7153]['time'] - pd.DateOffset(hours=1)
```

```python
# convert time to iso format
df['time'] = df['time'].dt.strftime("%Y-%m-%dT%H:%M:%SZ")
```

```python
# save to output file
df.to_csv(file_output, sep=',', index=False)
```
